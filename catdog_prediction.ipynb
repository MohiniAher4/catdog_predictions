```python
#cat=0
#dog=1
def predict(path):
    img=cv2.imread(path)
    img=cv2.resize(img,(224,224))/255
    y=model.predict_on_batch(img.reshape(1,224,224,3)).argmax()
    return "Cat" if y==0 else "Dog"
    
    
    
```
